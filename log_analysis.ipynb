```python
import re
from datetime import datetime
import matplotlib.pyplot as plt
from collections import Counter
```

```python
access_log = r"C:\Users\kenis\OneDrive\Desktop\3rd year 1st sem\Databases\Database-Project_FlavorFinder\access_log.unknown"
pattern = r'\[(\d{2}/\w{3}/\d{4}):(\d{2}):\d{2}:\d{2}'

times = []

with open(access_log, "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        match = re.search(pattern, line)
        if match:
            date, hour = match.groups()
            times.append(f"{date} {hour}:00")

count_by_time = Counter(times)

plt.figure(figsize=(8,4))
plt.bar(count_by_time.keys(), count_by_time.values())
plt.xticks(rotation=45)
plt.title("Website Access Timeline")
plt.xlabel("Time")
plt.ylabel("Number of Requests")
plt.tight_layout()
plt.savefig("access_timeline.png")
print("✅ access_timeline.png saved.")
```

```python
error_log = r"C:\Users\kenis\OneDrive\Desktop\3rd year 1st sem\Databases\Database-Project_FlavorFinder\error_log.unknown"
pattern_error = r'\[(\w{3} \w{3} \d{2} \d{2}:\d{2}:\d{2})'

errors = []

with open(error_log, "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        match = re.search(pattern_error, line)
        if match:
            errors.append(match.group(1))

count_errors = Counter(errors)

plt.figure(figsize=(8,4))
plt.bar(count_errors.keys(), count_errors.values(), color='red')
plt.xticks(rotation=90)
plt.title("Error Log Timeline")
plt.xlabel("Time")
plt.ylabel("Number of Errors")
plt.tight_layout()
plt.savefig("error_timeline.png")
print("✅ error_timeline.png saved.")
```
